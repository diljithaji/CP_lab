```python
import math

# Function to perform compare and swap
def compare_and_swap(arr, i, j, direction):
    """
    Compare two elements and swap depending on direction.
    direction = 1  → ascending
    direction = 0  → descending
    """
    if (direction == 1 and arr[i] > arr[j]) or (direction == 0 and arr[i] < arr[j]):
        arr[i], arr[j] = arr[j], arr[i]


# Bitonic merge function
def bitonic_merge(arr, low, count, direction):
    """
    Recursively merges bitonic sequence into sorted sequence.
    low       → starting index
    count     → number of elements
    direction → 1 for ascending, 0 for descending
    """
    if count > 1:
        k = count // 2

        # Compare elements at distance k
        for i in range(low, low + k):
            compare_and_swap(arr, i, i + k, direction)

        # Recursively merge first half
        bitonic_merge(arr, low, k, direction)

        # Recursively merge second half
        bitonic_merge(arr, low + k, k, direction)


# Bitonic sort recursive function
def bitonic_sort(arr, low, count, direction):
    """
    Creates bitonic sequence and sorts it.
    """
    if count > 1:
        k = count // 2

        # First half sorted ascending
        bitonic_sort(arr, low, k, 1)

        # Second half sorted descending
        bitonic_sort(arr, low + k, k, 0)

        # Merge entire sequence in given direction
        bitonic_merge(arr, low, count, direction)


# Utility function to pad array to next power of 2
def pad_to_power_of_two(arr):
    """
    Pads array with infinity to make its length a power of 2.
    """
    n = len(arr)
    next_power = 1 << (n - 1).bit_length()   # next power of 2

    padding_size = next_power - n
    arr.extend([float('inf')] * padding_size)

    return n   # return original length


# ================= MAIN EXECUTION =================

# Random array of 10 integers
arr = [29, 3, 17, 8, 25, 1, 40, 12, 9, 31]

print("Original Array:")
print(arr)

# Pad array
original_length = pad_to_power_of_two(arr)

# Perform bitonic sort (1 = ascending)
bitonic_sort(arr, 0, len(arr), 1)

# Remove padding
sorted_array = arr[:original_length]

print("\nSorted Array using Bitonic Sort:")
print(sorted_array)
```
